#  bmi
- input weight
- input height
- calculate bmi

```python
def bmi():
    name = input("Enter your name: ")
    height = float(input("Enter height in inches: "))*0.0254
    weight = float(input("Enter weight in kgs: "))
    bmi = weight/(height**2)
    print("Body mass index for {0} is: {1}".format(name.lower(),bmi))
```

# Data Visualization

```python
import seaborn as sns 
import matplotlib.pyplot as plt

sns.set_theme(style="ticks",color_codes=True)
```

```python
titanic = sns.load_dataset("titanic")
sns.catplot(x="sex",y="survived",hue="class",kind="bar",data=titanic)
plt.show()
```

```python
sns.catplot(x="sex",y="survived",hue="age",kind="bar",data=titanic)
plt.show()
```

```python
p1 = sns.countplot(x="sex",hue="survived",data=titanic)
p1.set_title("Plot for Counting")
plt.show()
```

```python
p1 = sns.countplot(x="sex",hue="class",data=titanic)
p1.set_title("Plot for Counting")
plt.show()
```

```python
g = sns.FacetGrid(row="sex",hue="alone",data=titanic)
g = (g.map(plt.scatter,"age","fare").add_legend())
```
