```python
# step 1 -- import libraries
import matplotlib.pyplot as plt
import seaborn as sns

# step 2 -- set a theme
sns.set_theme(style='ticks',color_codes=True)

# step 3 -- import dataset
titanic = sns.load_dataset('titanic')
# print(titanic)
```

```python
# Step-4-i plot basic graph with one variable(count)
p=sns.countplot(x='sex',data=titanic)
p.set_title("Titanic plot sex wise count plot")
plt.show()
```

```python
# Step-4-ii plot basic graph with two variable(count)
p=sns.countplot(x='sex',data=titanic, hue='class')
p.set_title("Titanic plot class wise sex count plot")
plt.show()
```
