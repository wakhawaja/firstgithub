# import Libraries
seaborn auto install these libraries
- *numpy*
- *scipy*
- *pandas*
- *matplotlib*

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flowers = sns.load_dataset('iris')
```

```python
flowers
```

## draw a lineplot

```python
sns.set_style(style=None, rc =None)
sns.set_style('white')
sns.lineplot(x='sepal_length',y='sepal_width',data=flowers)
plt.show()
```

## Adding titles

```python
sns.set_style(style=None, rc =None)
sns.set_style('dark')
sns.lineplot(x='sepal_length',y='sepal_width',data=flowers)
plt.title('Sepal Length to width plot')
plt.show()
```

## Adding Limits

```python
sns.set_style(style=None, rc =None)
sns.set_style('whitegrid')

sns.lineplot(x='sepal_length',y='sepal_width',data=flowers)
plt.title('Sepal Length to width plot')
plt.xlim(4.25)
plt.ylim(2.25)
plt.show()
```

## set setyles
- darkgrid
- whitegrid
- dark
- white
- ticks

```python
sns.set_style(style=None, rc =None)
sns.set_style('darkgrid')

sns.lineplot(x='sepal_length',y='sepal_width',data=flowers)
plt.title('Sepal Length to width plot')
plt.xlim(4.25)
plt.ylim(2.25)
plt.show()
```

## size of figure

```python
sns.set_style(style=None, rc =None)
sns.set_style('darkgrid')

#  change figure
plt.figure(figsize=(8,6))

sns.lineplot(x='sepal_length',y='sepal_width',data=flowers)
plt.title('Sepal Length to width plot')
plt.xlim(4.25)
plt.ylim(2.25)
plt.show()
```
